```python
# mengimport library, menggunakan numpy dan pandas
import numpy as np
import pandas as pd

# LabelEncoder untuk mengubah data menjadi nilai numerik
from sklearn.preprocessing import LabelEncoder
```

```python
# Memasukkan data ikan koi
ikan = pd.read_csv('./KOI2.csv', delimiter=';')

# Melihat 20 baris data
ikan.head(20)
```

```python
# Melihat Datanya
ikan
```

```python
# Menghapus Kolom Jumlah Corak,	Corak Dominan,	Spesifikasi
ikan.drop('Jumlah Corak', axis=1, inplace=True)
ikan.drop('Corak Dominan', axis=1, inplace=True)
ikan.drop('Tempat', axis=1, inplace=True)
```

```python
ikan
```

```python
# Mengubah data String menjadi Integer
encoder = LabelEncoder()
```

```python
# Merubah data String menjadi Angka
ikan ['Jenis Ikan'] = ikan['Jenis Ikan'].map({'Goromo':0, 'Kohaku':1, 'Sanke':2, 'Tancho':3, 'Shiro':4, 'Showa':5, 'Utsuri':6, 'Shushui':7, 'Chagoi':8, 'Platinum':9})
ikan ['Gender'] = ikan['Gender'].map({'Female':0})
ikan ['Umur'] = ikan['Umur'].map({'Dibawah 1 tahun':0, '1 - 3 Tahun':1, 'Diatas 3 Tahun':2})
ikan ['Spesifikasi'] = ikan['Spesifikasi'].map({'Local':0,'Import':1})
```

```python
ikan['Jenis Ikan'] = encoder.fit_transform(ikan['Jenis Ikan'])
ikan['Gender'] = encoder.fit_transform(ikan['Gender'])
ikan['Umur'] = encoder.fit_transform(ikan['Umur'])
ikan['Spesifikasi'] = encoder.fit_transform(ikan['Spesifikasi'])
```
